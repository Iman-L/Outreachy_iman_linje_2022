```python
import geopandas as gdp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
statesMWI = gdp.read_file("MWI_ALL_states.json")
```

## World Soil Groups

### Aim
- Find the world's soil groups 
- The distribution of soil groups in Malawi
- A study of soil groups in Malawi's four largest cities

### Significance
Agriculture can make use of this information by locating suitable lands for different types of crops.
### Findings 
- A total of 32 Major Soil Groups exist around the world.
- The soil in Malawi is predominantly ferralsol.
### Hypothesis 
Due to ferralsol soil's low fertility, Malawi uses fertilizers to increase food production.

```python
#Load world soil data
soil = gdp.read_file("World_Soil_Resources_wgs84.geojson")
```

```python
#Snippet of the data
soil.head()
```

```python
#Plot world soil map, colour coded. Notice number of soil groups. 
soil.plot(column=('IPCC'), figsize=(20,18), legend=True)
```

```python
#Load soil groups for Malawi
soil_M = gdp.read_file("MWI_AL2_Malawi_WSR.json")
#Show the data
soil_M.head()
```

```python
#Plot soil map of Malawi
soil_M.plot(column=('IPCC'), figsize=(9,15), legend=True)
plt.axis('off')
```

```python
#Import data of 4 majopr cities
soil_bt = gdp.read_file('MWI_AL4_Blantyre_WSR.json')
soil_Zomba = gdp.read_file('MWI_AL4_Zomba_WSR.json')
soil_Lilongwe = gdp.read_file('MWI_AL4_Lilongwe_WSR.json')
soil_Mzimba = gdp.read_file('MWI_AL4_Mzimba_WSR.json')

#Snippets
print(soil_bt, soil_Zomba,soil_Lilongwe,soil_Mzimba)
```

```python
#Plot soil map of Zomba and Blantyre, cities in the southern Malawi.
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(9, 8))
ax1 = soil_bt.plot(ax=ax1, column=('IPCC'), legend=True)
ax2 = soil_Zomba.plot(ax=ax2, column=('IPCC'), legend=True)
ax1.set_title("Blantyre City Soil Profile")
ax2.set_title("Zomba City Soil Profile")
ax1.axis('off')
ax2.axis('off')


#Plot soil map of Lilongwe and Mzimba, cities in Central and Northern region
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(10, 10))
ax1 = soil_Lilongwe.plot(ax=ax1, column=('IPCC'), legend=True)
ax2 = soil_Mzimba.plot(ax=ax2, column=('IPCC'), legend=True)
ax1.set_title("Lilongwe Capital City Soil Profile")
ax2.set_title("Mzimba City Soil Profile")
ax1.axis('off')
ax2.axis('off')
```
